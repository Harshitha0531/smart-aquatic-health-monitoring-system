## scaling and labelling

```python
max_ntu_df=df['Turbidity (NTU)'].max()
min_ntu_df=df['Turbidity (NTU)'].min()
```

```python
max_img_ntu=df_img['ntu'].max()
min_img_ntu=df_img['ntu'].min()
```

```python
print(max_img_ntu, min_img_ntu)
```

```python
print(max_ntu_df, min_ntu_df)
```

```python
for i, j in df.iterrows():
    df.loc[i, 'normalized_ntu'] = (df.loc[i, 'Turbidity (NTU)'] - min_ntu_df) / (max_ntu_df - min_ntu_df)
```

```python

for i, j in df.iterrows():
    df.loc[i, 'rescaled_ntu'] = min_img_ntu + (df.loc[i, 'normalized_ntu'] * (max_img_ntu - min_img_ntu))
```

```python
# Define class labels and ranges
turbidity_labels = {
    'Low turbidity': (340, 389),
    'Moderate turbidity': (390, 439),
    'High turbidity': (440, 489),
    'Very high turbidity': (490, 544)
}

# Function to label data
def label_data(value):
    for label, (min_val, max_val) in turbidity_labels.items():
        if min_val <= value <= max_val:
            return label
    return 'Unknown'

for i, j in df.iterrows():
    df.loc[i, 'turbidity_label']=label_data(df.loc[i, 'rescaled_ntu'])


```

```python
for i, j in df_img.iterrows():
    df_img.loc[i, 'ntu_label']=label_data(df_img.loc[i, 'ntu'])
```

```python
# Define class labels and ranges
DO_labels = {
    'Low': (6.0, 7.0),
    'Moderate': (7.1, 8.0),
    'High': (8.1, 9.0),
    'Very high': (9.1, 10.0)
}

# Function to label data
def label_data(value):
    for label, (min_val, max_val) in DO_labels.items():
        if min_val <= value <= max_val:
            return label
    return 'Unknown'

for i, j in df.iterrows():
    df.loc[i, 'DO_label']=label_data(df.loc[i, 'Dissolved Oxygen (mg/L)'])


```
